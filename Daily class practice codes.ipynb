```python
import pandas as pd
import numpy as np

# Function to generate synthetic student data
def generate_student_data(num_samples=1000):

    # Categories
    streams = ['CS', 'Mech', 'EC', 'IT', 'Civil']
    genders = ['F', 'M']

    # Random Seed
    np.random.seed(42)

    # Create DataFrame
    data = pd.DataFrame({
        'Stream': np.random.choice(streams, num_samples),
        'Gender': np.random.choice(genders, num_samples)
    })

    # Generate Marks (Scale: 1 to 10)
    data['Maths'] = np.round(np.clip(np.random.normal(6, 1, num_samples), 1, 10))
    data['English'] = np.round(np.clip(np.random.normal(7, 1, num_samples), 1, 10))
    data['Science'] = np.round(np.clip(np.random.normal(5, 1, num_samples), 1, 10))

    # Total Marks
    data['Total'] = data['Maths'] + data['English'] + data['Science']

    # Grade Assignment
    def assign_grade(total):
        if total >= 24:
            return 'A'
        elif total >= 18:
            return 'B'
        elif total >= 15:
            return 'C'
        else:
            return 'D'

    data['Grade'] = data['Total'].apply(assign_grade)

    # Fuel Consumption
    data['Fuel'] = np.round(
        np.clip(
            np.random.normal(10.5, 3.16, num_samples),
            1,
            20
        ),
        2
    )

    return data


# Generate Dataset
df = generate_student_data(1000)

# Display First Five Rows

print("FIRST FIVE ROWS")

print(df.head())

# Display Last Five Rows
print("\n")

print("LAST FIVE ROWS")

print(df.tail())

# Shape
print("\nDataset Shape")
print(df.shape)

# Columns
print("\nColumns")
print(df.columns)

# Information
print("\nDataset Information")
print(df.info())

# Statistics
print("\nStatistical Summary")
print(df.describe())

# Missing Values
print("\nMissing Values")
print(df.isnull().sum())

# Students in each Stream
print("\nStudents in each Stream")
print(df["Stream"].value_counts())

# Gender Count
print("\nGender Count")
print(df["Gender"].value_counts())

# Grade Count
print("\nGrade Distribution")
print(df["Grade"].value_counts())

# Average Marks
print("\nAverage Marks")
print("Maths :", df["Maths"].mean())
print("English :", df["English"].mean())
print("Science :", df["Science"].mean())

# Correlation
print("\nCorrelation Matrix")
print(df.corr(numeric_only=True))

# Grade A Students
print("\nTop Grade A Students")
print(df[df["Grade"] == "A"].head())

# Save CSV
df.to_csv("synthetic_student_data.csv", index=False)

print("\nDataset saved successfully as 'synthetic_student_data.csv'")
```

```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load Dataset

uploaded = files.upload()

filename = list(uploaded.keys())[0]

df = pd.read_csv(filename)

print("First 5 Rows")
print(df.head())

print("\nDataset Shape")
print(df.shape)

print("\nInformation")
print(df.info())

print("\nStatistics")
print(df.describe())

# Check Missing Values

print("\nMissing Values")
print(df.isnull().sum())

# Select Features

X = df[["energy_consumption", "temperature"]]

# Scale Data

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# Elbow Method

wcss = []

for i in range(1,11):

    model = KMeans(
        n_clusters=i,
        random_state=42,
        n_init=10
    )

    model.fit(X_scaled)

    wcss.append(model.inertia_)

# Plot Elbow Curve

plt.figure(figsize=(8,5))

plt.plot(range(1,11), wcss, marker='o')

plt.title("Elbow Method")

plt.xlabel("Number of Clusters")

plt.ylabel("WCSS")

plt.grid(True)

plt.show()

# Create Final Model

kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

df["Cluster"] = kmeans.fit_predict(X_scaled)

# Display Clustered Data

print("\nClustered Dataset")

print(df.head())

# Scatter Plot

plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="energy_consumption",
    y="temperature",
    hue="Cluster",
    palette="viridis"
)

plt.title("Energy Consumption Clusters")

plt.show()

# Cluster Summary

print("\nCluster Summary")

print(df.groupby("Cluster").mean(numeric_only=True))

# Number of Samples in Each Cluster

print("\nCluster Counts")

print(df["Cluster"].value_counts())

# Save Dataset

df.to_csv("clustered_energy_data.csv", index=False)

print("\nDataset saved successfully!")
```

```python
# ==========================================================
# K-Means Clustering - Energy Consumption Patterns
# ==========================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# ==========================================================
# Upload Dataset
# ==========================================================

uploaded = files.upload()

filename = list(uploaded.keys())[0]

df = pd.read_csv(filename)

print("Dataset Loaded Successfully!\n")

# ==========================================================
# Explore Dataset
# ==========================================================

print("First 5 Rows")
print(df.head())

print("\nShape of Dataset")
print(df.shape)

print("\nDataset Information")
print(df.info())

print("\nStatistical Summary")
print(df.describe())

print("\nMissing Values")
print(df.isnull().sum())

# ==========================================================
# Handle Missing Values
# ==========================================================

df.fillna(df.mean(numeric_only=True), inplace=True)

# ==========================================================
# Select Features
# ==========================================================

X = df[['energy_consumption', 'temperature']]

# ==========================================================
# Feature Scaling
# ==========================================================

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# ==========================================================
# Elbow Method
# ==========================================================

wcss = []

for i in range(1, 11):

    kmeans = KMeans(
        n_clusters=i,
        init='k-means++',
        random_state=42,
        n_init=10
    )

    kmeans.fit(X_scaled)

    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))

plt.plot(range(1,11), wcss, marker='o')

plt.title("Elbow Method")

plt.xlabel("Number of Clusters")

plt.ylabel("WCSS")

plt.grid(True)

plt.show()

# ==========================================================
# Final Model
# ==========================================================

best_k = 3

kmeans = KMeans(
    n_clusters=best_k,
    init='k-means++',
    random_state=42,
    n_init=10
)

df['cluster'] = kmeans.fit_predict(X_scaled)

# ==========================================================
# Evaluate Model
# ==========================================================

silhouette_avg = silhouette_score(X_scaled, df['cluster'])

print("\nSilhouette Score :", round(silhouette_avg,3))

# ==========================================================
# Cluster Summary
# ==========================================================

print("\nCluster Summary")

print(df.groupby('cluster').mean(numeric_only=True))

print("\nCluster Counts")

print(df['cluster'].value_counts())

# ==========================================================
# Visualisation
# ==========================================================

plt.figure(figsize=(10,6))

sns.scatterplot(
    x='energy_consumption',
    y='temperature',
    hue='cluster',
    data=df,
    palette='viridis'
)

plt.title('Clusters of Energy Consumption Patterns')

plt.xlabel('Energy Consumption')

plt.ylabel('Temperature')

plt.legend(title='Cluster')

plt.show()

# ==========================================================
# Save Dataset
# ==========================================================

df.to_csv("clustered_energy_data.csv", index=False)

print("\nClustered dataset saved successfully!")
```

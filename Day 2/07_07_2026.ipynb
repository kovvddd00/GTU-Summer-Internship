```python
# 1. LIST (5 METHODS)
print("\n1. LIST METHODS")

fruits = ["Apple", "Banana", "Mango"]
print("Original List:", fruits)

fruits.append("Orange")
print("append():", fruits)

fruits.insert(1, "Kiwi")
print("insert():", fruits)

fruits.remove("Banana")
print("remove():", fruits)

fruits.pop()
print("pop():", fruits)

fruits.sort()
print("sort():", fruits)


# 2. TUPLE (5 OPERATIONS)
print("\n2. TUPLE")

numbers = (10, 20, 30, 20, 50)

print("Tuple:", numbers)
print("count():", numbers.count(20))
print("index():", numbers.index(30))
print("len():", len(numbers))
print("max():", max(numbers))
print("min():", min(numbers))


# 3. SET (5 METHODS)
print("\n3. SET METHODS")

colors = {"Red", "Blue", "Green"}
print("Original Set:", colors)

colors.add("Yellow")
print("add():", colors)

colors.update(["Black", "White"])
print("update():", colors)

colors.remove("Blue")
print("remove():", colors)

colors.discard("Pink")
print("discard():", colors)

removed = colors.pop()
print("pop(): Removed ->", removed)
print("Remaining Set:", colors)


# 4. DICTIONARY (5 METHODS)
print("\n4. DICTIONARY METHODS")

student = {
    "Name": "Rahul",
    "Age": 20,
    "Marks": 90
}

print("Dictionary:", student)

print("keys():", student.keys())
print("values():", student.values())
print("items():", student.items())

student.update({"City": "Surat"})
print("update():", student)

student.pop("Age")
print("pop():", student)


# 5. IF
print("\n5. IF STATEMENT")

age = 20

if age >= 18:
    print("Eligible to Vote")


# 6. IF ELSE
print("\n6. IF ELSE")

marks = 35

if marks >= 40:
    print("Pass")
else:
    print("Fail")


# 7. IF ELIF ELSE
print("\n7. IF ELIF ELSE")

marks = 82

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")


# 8. NESTED IF ELSE
print("\n8. NESTED IF ELSE")

age = 21
citizen = True

if age >= 18:
    if citizen:
        print("Eligible to Vote")
    else:
        print("Not an Indian Citizen")
else:
    print("Not Eligible")


# 9. BREAK
print("\n9. BREAK")

for i in range(1, 11):
    if i == 6:
        break
    print(i)


# 10. CONTINUE
print("\n10. CONTINUE")

for i in range(1, 11):
    if i == 5:
        continue
    print(i)


# 11. PASS
print("\n11. PASS")

for i in range(5):
    if i == 2:
        pass
    print(i)


# 12. INPUT FUNCTION
print("\n12. INPUT FUNCTION")

name = input("Enter your name: ")

print("Welcome", name)


# 13. RANGE FUNCTION
print("\n13. RANGE FUNCTION")

for i in range(1, 6):
    print(i)


# 14. LEN FUNCTION
print("\n14. LEN FUNCTION")

text = "Python Programming"

print("Length =", len(text))


# 15. TYPE FUNCTION
print("\n15. TYPE FUNCTION")

a = 10
b = 15.75
c = "Python"
d = True

print(type(a))
print(type(b))
print(type(c))
print(type(d))


# 16. FOR LOOP
print("\n16. FOR LOOP")

for i in range(1, 6):
    print("Iteration:", i)


# 17. WHILE LOOP
print("\n17. WHILE LOOP")

i = 1

while i <= 5:
    print("Count:", i)
    i += 1
```

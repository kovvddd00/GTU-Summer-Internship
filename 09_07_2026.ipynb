```python
import pandas as pd
import numpy as np

from google.colab import files

uploaded = files.upload()

filename = list(uploaded.keys())[0]

df = pd.read_csv(filename)

print("=" * 70)
print("GREEN SKILLING & SUSTAINABILITY DATA ANALYSIS")
print("=" * 70)

print("\nDataset Preview")
print(df.head())

print("\nDataset Shape")
print(df.shape)

print("\nColumn Names")
print(df.columns.tolist())

print("\nDataset Information")
print(df.info())

print("\nStatistical Summary")
print(df.describe())

possible_columns = [
    "country",
    "year",
    "population",
    "gdp",
    "co2",
    "renewables_share_energy",
    "solar_electricity",
    "wind_electricity",
    "hydro_electricity",
    "fossil_fuel_consumption"
]

available_columns = [col for col in possible_columns if col in df.columns]

green_df = df[available_columns].copy()

print("\nSelected Columns")
print(green_df.head())

print("\nMissing Values")
print(green_df.isnull().sum())

print("\nMissing Value Percentage")
missing = (green_df.isnull().sum() / len(green_df)) * 100
print(missing)

numeric_columns = green_df.select_dtypes(include=np.number).columns
green_df[numeric_columns] = green_df[numeric_columns].fillna(green_df[numeric_columns].median())

categorical_columns = green_df.select_dtypes(include="object").columns
green_df[categorical_columns] = green_df[categorical_columns].fillna("Unknown")

print("\nMissing Values After Cleaning")
print(green_df.isnull().sum())

print("\nCorrelation Matrix")
print(green_df.corr(numeric_only=True))

if "renewables_share_energy" in green_df.columns:
    print("\nTop 10 Renewable Energy Share")
    print(
        green_df.sort_values(
            by="renewables_share_energy",
            ascending=False
        ).head(10)
    )

if "co2" in green_df.columns:
    print("\nTop 10 CO₂ Emissions")
    print(
        green_df.sort_values(
            by="co2",
            ascending=False
        ).head(10)
    )

if "country" in green_df.columns:
    print("\nNumber of Countries")
    print(green_df["country"].nunique())

if "year" in green_df.columns:
    print("\nYear Range")
    print("Minimum Year :", green_df["year"].min())
    print("Maximum Year :", green_df["year"].max())

print("\nCleaned Dataset Preview")
print(green_df.head())
```
